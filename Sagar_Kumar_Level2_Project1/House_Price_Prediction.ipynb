# Project 5: Predicting House Prices with Linear Regression

**Dataset:** `dataset/Housing.csv` (target: `price`)

**Goals:** Explore and clean data, select features, train a linear regression model, evaluate with MSE and R-squared, and visualize predictions.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from IPython.display import Image, display

import house_price_regression as hp

sns.set_theme(style='whitegrid', font_scale=1.05)
hp.FIG_DIR.mkdir(parents=True, exist_ok=True)
```

## 1. Data Collection and Exploration

```python
df = hp.load_data()
df = hp.explore_and_clean(df)
df.head()
```

## 2. Feature Selection

```python
numeric_features, categorical_features, corr = hp.select_features(df)
corr
```

## 3. Exploratory Visualizations

```python
hp.plot_correlation_heatmap(df)
hp.plot_price_distribution(df)
hp.plot_area_vs_price(df)
for f in ['01_correlation_heatmap.png', '02_price_distribution.png', '03_area_vs_price.png']:
    display(Image(filename=str(hp.FIG_DIR / f)))
```

## 4. Model Training (Scikit-Learn Linear Regression)

```python
model, y_train, y_test, y_pred_train, y_pred_test = hp.train_model(
    df, numeric_features, categorical_features
)
```

## 5. Model Evaluation (MSE, R-squared)

```python
train_metrics = hp.evaluate_model(y_train, y_pred_train, 'Training')
test_metrics = hp.evaluate_model(y_test, y_pred_test, 'Test')
pd.DataFrame([{'split': 'train', **train_metrics}, {'split': 'test', **test_metrics}])
```

## 6. Predicted vs Actual Visualization

```python
hp.plot_actual_vs_predicted(y_test, y_pred_test, 'Test', '04_actual_vs_predicted_test.png')
hp.plot_residuals(y_test, y_pred_test)
hp.save_coefficients(model, numeric_features, categorical_features)
display(Image(filename=str(hp.FIG_DIR / '04_actual_vs_predicted_test.png')))
display(Image(filename=str(hp.FIG_DIR / '05_residual_analysis.png')))
```
