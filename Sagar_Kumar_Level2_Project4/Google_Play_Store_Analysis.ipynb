# Project 8: Google Play Store Analysis

**Datasets:** `dataset/apps.csv`, `dataset/user_reviews.csv`

**Goals:** Clean data, explore categories, analyze ratings/pricing/size, sentiment analysis, interactive visualizations.


```python
import pandas as pd
from IPython.display import Image, display, IFrame
import google_play_analysis as gp
```

## 1. Data Preparation

```python
apps = gp.clean_apps(gp.load_apps())
reviews = gp.clean_reviews(gp.load_reviews())
merged = gp.merge_apps_reviews(apps, reviews)
apps.head()
```

## 2. Category Exploration

```python
cat_summary = gp.category_exploration(apps)
cat_summary.head(15)
```

## 3. Metrics Analysis (Ratings, Pricing, Size)

```python
stats = gp.metrics_analysis(apps)
pd.DataFrame([stats])
```

## 4. Sentiment Analysis

```python
sentiment_by_cat = gp.sentiment_analysis(reviews, merged)
sentiment_by_cat.head(10)
```

## 5. Static Visualizations

```python
gp.plot_category_distribution(apps)
gp.plot_rating_distribution(apps)
gp.plot_price_analysis(apps)
gp.plot_size_vs_rating(apps)
gp.plot_installs_vs_rating(apps)
gp.plot_sentiment_distribution(reviews)
gp.plot_sentiment_by_top_categories(merged)
for f in sorted(gp.FIG_DIR.glob('*.png')):
    print(f.name)
    display(Image(filename=str(f)))
```
